**Question 17:** Write a Python program to check if a year is a leap year.

```python
## A year is a leap year if it is divisible by 4, except year divisible by 100 unless it is divisible by 400.

year = 2000
if year%4 == 0:
    if year%100 == 0:
        if year%400 == 0:
            print(year,"is a leap year")
        else:
            print(year,"is not a leap year")
    else:
        print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```
