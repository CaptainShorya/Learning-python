```python
##Assignment 13
##defaultdict(list) → Every new key automatically gets an empty list []
from collections import defaultdict
d = defaultdict(list) ## not defaultdict([])
d["a"].append(1)
d["b"].append(2)
print(d)
##A defaultdict with list as default value automatically initializes each new key with an empty list, allowing direct appending of elements.
```
