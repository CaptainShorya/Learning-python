```python
##Assignment 7
with open('reverse.txt','r') as file:
    just = file.readlines() 
    print(reversed(just)) ## returns a reverse iterator,not list of strings
    for line in reversed(just): ##Starts from the last line,Moves backward to the first line
        print(line.strip()) ##strip() removes: newline,leading/trailing spaces
```

```python
##Assignment 8 -> Great problem to build strong basics
with open('document.txt','r') as file:
    lines = file.readlines() ## create list of lines ['line1','line2']
    print(lines)
    word_count = 0
    line_count = 0
    char_count = 0
    for line in lines:
        print(line.strip()) ## Pick line one by one from list of line
        print(line.split()) ## form a list of words
        line_count = line_count + 1
        word_count = word_count + len(line.split()) ## summation of length of particular list
        char_count = char_count + len(line)
    print(line_count)
    print(word_count)
    print(char_count)
```

```python
##Assignment 13
import csv
with open('data.csv','w') as file:
    write = csv.writer(file) ##created write object to write inside file
    write.writerows([["Alice", 25, "Delhi"],["BOB", 29, "Bombay"]]) ## takes list of multiple(nested) list
    write.writerow(["JOhn", 53, "Chennai"])
with open('data.csv','r') as file:
    read = csv.DictReader(file)
    print(read) ## return csv.DictReader Object so better convert it into list to see visiblity of data properly
    print(list(read))
```

```python
##Assignment 14
import json
dict1 = {1:"Hero",2:"TVS"}
with open('data.json', 'w') as file:
    #Format 1
    #data = json.dump(dict1,file)
    #print(data,type(data))
    #Format 2
    data = json.dumps(dict1) ##json.dumps(dict1) only converts the dictionary to a JSON string
    file.write(data)
    print(data,type(data))
with open('data.json', 'r') as file:
    data = json.load(file) ##
    print(data,type(data)) 

##Learning - > dump/load → files, dumps/loads → strings
##json.dumps() =>	dict → JSON string
##json.dump()  =>   dict → JSON file
##json.loads() =>  JSON string → dict
##json.load()  =>  JSON file → dict
```
