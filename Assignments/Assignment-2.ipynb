```python
##Assignment 15 (Fibonacci Number)
n = int(input("Enter a number: "))
n1 = 0
print(n1,end =" ")
n2 = 1
for i in range(1,n):
    print(n2,end = " ")
    sum = n1 + n2
    n1 = n2
    n2 = sum
```
