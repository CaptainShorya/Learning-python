```python
##Assignment 4
class Bank_account:
    def __init__(self,account_name,balance=0):
        self.__account_name = account_name
        self.__balance = balance
    
    def deposit(self,amount):
        self.__balance += amount
        print(f'{amount} deposited and current balance is {self.__balance}')

    def withdraw(self,amount):
        if amount > self.__balance:
            print("Insufficient Balance!!")
        else:
            self.__balance -= amount
            print(f'{amount} withdraw and current balance is {self.__balance}')

    def check_balance(self):
        print(f'Current balance is : {self.__balance}')

account1 = Bank_account("Shorya",10000)
account1.deposit(2000)
account1.withdraw(1000)

account1.check_balance()
account1.withdraw(12000)
        
```

```python
##Assignment 8
class Counter:
    count = 0 #Class variable → shared by all objects
    def __init__(self):
        Counter.count += 1 # Always access shared data via the class name.

    @classmethod # The count belongs to the class, not a single object
    def get_count(cls): #cls refers to the class (Counter)
        return cls.count

#A class method is used to return the current count because the variable tracking the number of objects is a class-level attribute shared by all instances.

c1 = Counter()
print(c1.get_count())
c2 = Counter()
print(c2.count)
```

```python
## Assignment 9
import math 
#static Method -> It does not use self, It does not use class data (cls), It behaves like a normal function but is grouped inside a class
class MathOperations:
    @staticmethod #What @staticmethod does: tells Python: “This method does not need self”
    def sqrt(x):
        return math.sqrt(x)
    
rt = MathOperations()
print(rt.sqrt(16))
```

```python
## Assignment 10
class Rectangle:
    def __init__(self,length,width):
        self.__length = length
        self.__width = width
    
    # Getters should RETURN, not PRINT
    def get_length(self):
        return self.__length
    def get_width(self):
        return self.__width

    def set_length(self,length):
        self.__length = length
    def set_width(self,width):
        self.__width = width

rec1 = Rectangle(10,5)
print(rec1.get_length())
print(rec1.get_width())

rec1.set_length(100)
rec1.set_width(500)
print(rec1.get_length(),rec1.get_width())

```

```python
##Assignment 11
from abc import ABC,abstractmethod

class Area(ABC): # Abstract class - The class whose we never create an object
    @abstractmethod
    def area(self):
        pass

class Circle(Area):
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return 3.14*self.radius*self.radius

class Square(Area):
    def __init__(self,side):
        self.side = side
    def area(self):
        return self.side*self.side

sq = Square(10)
print(sq.area())
cr = Circle(5)
print(cr.area())
```

```python
##Assignment 12
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

##Operator Overloading -> Giving a new meaning to an operator (+, -, *, etc.) for user-defined objects.
## + → calls __add__
## v1 + v2 becomes -> v1.__add__(v2)

    def __add__(self,other):
        return Vector(self.x + other.x,self.x + other.y)

## self → first vector (v1), other → second vector (v2) and we create and return a new value
    def __str__(self):
        return f'Vector({self.x}, {self.y})'
##define __str__ for readability.

v1 = Vector(2,3)
v2 = Vector(4,5)

v3 = v1 + v2 ## Operator Overloading
print(v3)

```

```python
##Assignment 13
class InsufficientBalanceError(Exception): #You are creating a new exception type, It inherits from Exception
    pass

class BankAccount:
    def __init__(self, account_number, balance=0):
        self.__account_number = account_number
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount > self.__balance:
            raise InsufficientBalanceError("Insufficient balance!")
        # raise does: Immediately stops the method, creates an exception object, Control jumps to the nearest except block, Stores error in e, Prints the message
        else:
            self.__balance -= amount

    def check_balance(self):
        return self.__balance
    
account = BankAccount('12345678', 1000)
account.deposit(500)

try:
    account.withdraw(2000)
except InsufficientBalanceError as e:
    print(f"Error: {e}")
```

```python
##Assignment 15
## Method chaining works by returning the object itself from each method, allowing multiple method calls on the same object in a single expression.
## In short, Method chaining = return self from every method.
class Calculator:
    def __init__(self, value=0):
        self.value = value

    def add(self, amount):
        self.value += amount
        return self

    def subtract(self, amount):
        self.value -= amount
        return self

    def multiply(self, amount):
        self.value *= amount
        return self

    def divide(self, amount):
        if amount != 0:
            self.value /= amount
        else:
            print("Cannot divide by zero!")
        return self
    
calc = Calculator()
calc.add(10).subtract(3).multiply(2).divide(2)
print(calc.value)  

#add(10) → value = 10 → returns self
#multiply(2) → value = 20 → returns self
#subtract(5) → value = 15 → returns self
#divide(5) → value = 3 → returns self
```
